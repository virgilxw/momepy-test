```python
import geopandas as gpd
import pandas as pd
import pickle
import momepy
import dask
from dask.distributed import Client, LocalCluster
from tqdm.auto import tqdm
```

```python
daskCluster = LocalCluster(threads_per_worker=2,
                n_workers=8, memory_limit='100GB')

client = Client(daskCluster, timeout="300s")

client
```

```python
tessellation = gpd.read_parquet("./out/singapore/tessellation_stats.pq")
buildings = gpd.read_parquet("./out/singapore/buildings.pq")

with open('./out/singapore/weights.pickle', 'rb') as f:
    queen_1 = pickle.load(f)
    queen_3 = pickle.load(f)
    dist200 = pickle.load(f)
    f.close()
```

```python
tessellation_dropped = tessellation.drop(["geometry", "uID"], axis=1)
```

```python
chunk_size = 16

chunks = [tessellation_dropped.columns[i:min(i+chunk_size, len(tessellation_dropped.columns))] for i in range(0, len(tessellation_dropped.columns), chunk_size)]

len_sublist = len(chunks)

```

```python
percentiles = []
delayed_obj = []
tessellation_scatter = client.scatter(tessellation)
queen_3_scatter = client.scatter(queen_3)

results = []

for i, sublist in enumerate(chunks):

    delay_objs = [dask.delayed(momepy.Percentiles)(tessellation_scatter, column, queen_3_scatter, "uID") for column in tqdm(sublist, desc=f"Computing chunk {i+1}/{len_sublist}")]
    new_results = dask.compute(*delay_objs)
    {results.append(result) for result in new_results}
    
    
# for column in tqdm(tessellation_dropped.columns, total=len(tessellation_dropped.columns)):
#     delayed_obj.append(dask.delayed(momepy.Percentiles)(tessellation_scatter, column, queen_3_scatter, "uID"))

    # .frame
# perc.columns = [f"{column}_" + str(x) for x in perc.columns]
# percentiles.append(perc)
```
